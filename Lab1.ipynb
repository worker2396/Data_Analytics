```python
stan_model = cmdstanpy.CmdStanModel(stan_file='bern_1.stan')
print(stan_model)
```

```python
fit = stan_model.sample(data = dicts)
```

```python
theta = fit.stan_variable('theta')
df=fit.summary()
df_theta = df.loc['theta']
q5 = df_theta['5%']
median = df_theta['50%']
q95 = df_theta['95%']
df
```

```python
plt.hist(theta,bins=50,density=True)
plt.axvline(q5,color='r')
plt.axvline(theta.mean(),color='y')
plt.axvline(median,color='g')
plt.axvline(q95,color='brown')
plt.show()
```
